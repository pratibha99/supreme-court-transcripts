Topic Modeling
--

Create a dictionary that contains label as 10 categories of cases, for example, property, relationships, trade licenses, content as the corresponding key words, for example, "child","mother" for the category "relationship".  Macth each case with the 10 categories to see the frequencies of keywords in each case. The category with the most hitted keyword is the label of the case.

The dictionary comes from a research "READING THE HIGH COURT AT A DISTANCE: TOPIC MODELLING THE LEGAL SUBJECT MATTER AND JUDICIAL ACTIVITY OF THE HIGH COURT OF AUSTRALIA, 1903–2015", conducted by UNSW Law School. The dictionary was built through an iterative process, first by having intuitive labels based on the dominant words in each topic, and refined by analysing the cases that is categories as certain topic based on the greatest alignment with the topic. 

Topic = [G, B, Tr, Tra, D, C, J, E, Con]

G = "Government action in relation to immigration"

B = 'Beneficiaries’ rights to property in an estate'

Tr = 'Trial process for criminal matters'

Tra = 'Trade licencing, regulation and IP'

D = 'Damage to persons through injury'

C = 'Company financial flows'

J = 'Jurisdictional divisions and actions'

E = 'Employment entitlements and disputes'

L = 'Land contracts and agreements'

Con = 'Constitutional actors and relationships'

```python
k = k.assign(G = 0, B = 0, Tr = 0, Tra = 0, D = 0, C = 0, J = 0, E = 0, L = 0, Con = 0)
Topic = ['G', 'B', 'Tr', 'Tra', 'D','C', 'J', 'E', 'L', 'Con']

threshold = 10

for index, row in k.iterrows():
    for i in Topic: 
        if (myDict[row['words']] == i and row['frequencies'] > threshold):
            k.at[index, i] = row['frequencies']
k
```

```python
new_k = k[['Case','G','B','Tr','Tra','D','C','E','J','L','Con']]
new_k = new_k.pivot_table(new_k, index=["Case"], aggfunc=np.sum)
new_k = new_k.reset_index(level=['Case'])
```

```python
no_case = new_k.drop(['Case'], axis=1)
new_k['Topic'] = no_case.idxmax(axis=1)
new_k
```
