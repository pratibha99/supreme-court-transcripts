```python
import pandas as pd
import regex as re
import numpy as np
import os
from os import listdir
from os.path import isfile,join
```

```python
files = []
path = "/Users/yuhanxie/Desktop/LA/textfiles/"
for i in os.listdir(path):
    if i.endswith('.txt'):
        text = open(path+i,encoding="utf8").read()
        files.append(text)
```

```python
def unique(texts):
    list(set(texts))
#return a set of nonunique bill mentioned in the speech.
def bill(texts):
    no_num = re.sub('\d+','',texts)
    no_n = re.sub('\n','',no_num)
    evenly_spaced = re.sub('\s+',' ',no_n)
    r = re.findall(r'(([A-Z][a-z]+\s)+Act)', evenly_spaced)
    b = []
    for i in range(len(r)):
        b.append(r[i][0])
    return b

#return a set of unique bill mentioned in the speech.
def unibill(texts):
    return np.unique(bill(texts))
#return the name and times it appears.
def count(texts):
    bills = bill(texts)
    ret = []
    for i in range(0,len(bills)):
        count = bills.count(bills[i])
        ret.append (str(bills[i]) + " " + ":" + ' '+str(count))
    return unique(ret)
```

```python
cleaned = []
for txt in files:
    clean = re.sub('\xad','',txt)
    clean = re.sub('\n','',clean)
    clean = re.sub('\\\\','',clean)
    cleaned += [clean]
```
